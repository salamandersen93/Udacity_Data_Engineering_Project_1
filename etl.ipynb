```python
def get_files(filepath):
    all_files = []
    for root, dirs, files in os.walk(filepath):
        files = glob.glob(os.path.join(root,'*.json'))
        for f in files :
            all_files.append(os.path.abspath(f))
    
    return all_files
```

## #3: `time` Table
#### Extract Data for Time Table
- Filter records by `NextSong` action
- Convert the `ts` timestamp column to datetime
  - Hint: the current timestamp is in milliseconds
- Extract the timestamp, hour, day, week of year, month, year, and weekday from the `ts` column and set `time_data` to a list containing these values in order
  - Hint: use pandas' [`dt` attribute](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.dt.html) to access easily datetimelike properties.
- Specify labels for these columns and set to `column_labels`
- Create a dataframe, `time_df,` containing the time data for this file by combining `column_labels` and `time_data` into a dictionary and converting this into a dataframe

```python
df_ns = df[df['page'] == 'NextSong']
print(df_ns.shape)
df_ns.head()
```

```python
t = df_ns.copy()
t['ts'] = pd.to_datetime(df_ns['ts'], unit='ms')
t.head()
```

```python
t['year'] = pd.DatetimeIndex(t['ts']).year
t['month'] = pd.DatetimeIndex(t['ts']).month
t['week'] = pd.DatetimeIndex(t['ts']).week
t['day'] = pd.DatetimeIndex(t['ts']).day
t['weekday'] = pd.DatetimeIndex(t['ts']).weekday
t['hour'] = pd.DatetimeIndex(t['ts']).hour


# converting df to dict
df_dict = t.T.to_dict()

list_of_time_data = []

# only want dates
for key, value in df_dict.items():
    entry = value
    dates_dict = {k: v for k, v in entry.items() if k in ['year', 'month', 'week', 'day', 'weekday', 'hour', 'ts']}
    list_of_time_data.append(dates_dict)

for i in range(len(list_of_time_data)):
    print(list_of_time_data[i])
    if i > 5: break # keeping the view clean
```

```python
time_df = pd.DataFrame(list_of_time_data)

# want order consistent with SQL INSERT query
time_df = time_df[['ts', 'hour', 'day', 'week', 'month', 'year', 'weekday']]

time_df.head()
```

#### Insert Records into Time Table
Implement the `time_table_insert` query in `sql_queries.py` and run the cell below to insert records for the timestamps in this log file into the `time` table.
```python
for i, row in time_df.iterrows():
    cur.execute(time_table_insert, list(row))
    conn.commit()
```
